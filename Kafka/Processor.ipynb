```python
import requests
import json
import time
from kafka import KafkaConsumer
from kafka import KafkaProducer
# #from config import *
import time
import datetime
from datetime import date
from datetime import timedelta
import os
import statistics

topic_name = 'coin'
servers = 'localhost:9092'
LIMIT = 30
N = 5
```

```python
consumer = KafkaConsumer(bootstrap_servers='localhost:9092')
consumer.subscribe(topics=['coin'])
producer = KafkaProducer(bootstrap_servers='localhost:9092', api_version=(0, 11, 5))
```

```python
## create volume dictionary

coint_list = [
    'BTC', 'BNB', 'ETH', 'DOGE', 'XRP', 'SHIB',
    'ADA', 'ARB', 'MATIC', 'SOL', 'PEPE'
]

coin_volume_dict = {coint_list[i]: [] for i in range(len(coint_list))}
```

```python
for message in consumer:
    
    coin = json.loads(message.value.decode('utf-8'))['name']
    volume = json.loads(message.value.decode('utf-8'))['volume']
    order = json.loads(message.value.decode('utf-8'))
    coin_volume_dict[coin].append(volume)
    
    if len(coin_volume_dict[coin]) == 11:
        if volume > statistics.mean(coin_volume_dict[coin]):
            order['action'] = 'buy'
        else:
            order['action'] = 'ignore'
        producer.send('buying_orders', json.dumps(order).encode('utf-8'))
        coin_volume_dict[coin] = coin_volume_dict[coin][1:]
    print(order, len(coin_volume_dict[coin]))
```
